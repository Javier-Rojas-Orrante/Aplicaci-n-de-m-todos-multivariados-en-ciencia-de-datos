```python
for columna in centro.columns:
  print(f"Columna: {columna}")

  q1 = centro[columna].quantile(0.25)
  q3 = centro[columna].quantile(0.75)
  iqr = q3 - q1
  abajo = q1 - 1.5 * iqr
  arriba = q3 + 1.5 * iqr

  outliers = centro[(centro[columna] < abajo) | (centro[columna] > arriba)][columna]
  print(f"  Outliers (IQR): {len(outliers)} valores")

  if len(outliers) > 0:
    print(f"  Rango de outliers: [{outliers.min():.2f}, {outliers.max():.2f}]")
```

```python
bC = centro.boxplot(figsize=(15, 10))
```

```python
centro1 = centro.drop_duplicates()
centro1.shape
```

```python
suroeste1 = suroeste.drop_duplicates()
suroeste1.shape
```

```python
sur1 = sur.drop_duplicates()
sur1.shape
```

```python
from scipy.stats.mstats import winsorize

centroW = centro1.copy()
for columna in centroW.columns:
  centroW[columna] = winsorize(centroW[columna], (0.1, 0.1))

centroW
```

```python
suroesteW = suroeste1.copy()
for columna in suroesteW.columns:
  suroesteW[columna] = winsorize(suroesteW[columna], (0.1, 0.1))

suroesteW
```

```python
surW = sur1.copy()
for columna in surW.columns:
  surW[columna] = winsorize(surW[columna], (0.1, 0.1))

surW
```
